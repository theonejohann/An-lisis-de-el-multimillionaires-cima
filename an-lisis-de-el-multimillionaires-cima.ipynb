## 1. Visualizaciones de el Datos
---

### 1.1 Grafico de Bar
---

```python
def barplt(year):
    plt.figure(figsize=(20,10))
    plt.title("Mutimillionaires cima en "+ year)
    plt.bar(df.loc[:, "NAME"].tolist(), df.loc[: ,year].tolist())

barplt("2018")
```

### 1.2 Grafico de línea
---

```python
def one_value_linegraph(value):
    
    plt.title(value + " Net worth")
    plt.xlabel("Ano")
    plt.ylabel("Dinero")
    y = df.loc[df['NAME'] == value].values.tolist()[0]
    del y[0]
    plt.plot( [2018, 2019, 2020, 2021, 2022], y)

    
one_value_linegraph("Elon Musk")
```


### 1.3 Grafico de Pie
---

```python
def piechart_all(year):
    names = df["NAME"].tolist()
    money = df[year].tolist()
    plt.title("Todos Mutimillionaires Cuota en " + year)
    plt.pie(money, labels=names, autopct='%1.1f%%')
```

```python
piechart_all("2018")
```
